```python
# Importamos las librerías necesarias

import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#EXPLORACION DE CSV

df_uber = pd.read_csv("rideshare_kaggle.csv")

display(df_uber.head(2))

print("=" * 100)

display(df_uber.tail(2))

print("=" * 100)

display(df_uber.sample(2))
```

```python
print(f"El dataframe UBER tiene {df_uber.shape[0]} filas y {df_uber.shape[1]} columnas")
```

```python
resumen_nulos = pd.DataFrame({
    'Tipo de dato': df_uber.dtypes,
    'No nulos conteo': df_uber.count(),
    'Nulos conteo': df_uber.isnull().sum(),
    'Porcentaje de nulos (%)': (df_uber.isnull().sum() / len(df_uber)) * 100
}).round(2)
display(resumen_nulos)
```

# LIMPIEZA E IMPUTACIÓN DE NULOS 

```python
df_uber["price"].describe()
```

```python
media = df_uber["price"].mean()
mediana = df_uber["price"].median()

sns.boxplot(x=df_uber["price"])

plt.axvline(media, color="purple", linestyle="--", label="Media")
plt.axvline(mediana, color="yellow", linestyle="--", label="Mediana")

plt.legend()
plt.show()
```

```python
df_limpio = df_uber[df_uber["price"].notnull()]
df_limpio = df_uber[df_uber["price"] > 0]
```

```python
print(df_limpio["price"].isnull().sum())
```

```python
#para ver los duplicados
print(df_limpio.duplicated().sum())
```
